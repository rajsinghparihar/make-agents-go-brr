```python
from crewai_tools import NL2SQLTool
from crewai.project import agent, CrewBase
from crewai import Agent, Task, Crew, LLM
from dotenv import load_dotenv
import os
```

```python
load_dotenv(".env", override=True)
```

```python
# psycopg2 was installed to run this example with PostgreSQL
nl2sql = NL2SQLTool(db_uri="sqlite:///data/electronics_transaction.db")
```

```python
llm = LLM(model="groq/gemma2-9b-it", temperature=0.3)
```

```python
agents_config = {
    "researcher": {
        "role": "Senior Data Scientist",
        "goal": "Expert in data modeling and SQL",
        "backstory":(
            "You're a seasoned data scientist and data engineer with a knack for data modeling and analysis. "
            "Known for your ability to find the most relevant "
            "information and present it in a clear and concise manner. using tools like SQL, you are able to write and execute complex queries on large databases."
        ) 
    }
}
```

```python
import ast
```

```python
ast.literal_eval("['iPhone 15', 'iPhone 15 Pro Max', 'iPhone 15 Pro', 'Galaxy A15 4G', 'Galaxy A15 5G', 'Galaxy A35 5G', 'iPhone 14', 'iPhone 14 Pro', 'iPhone 14 Pro Max', 'Galaxy S23']")
```

```python
@agent
def researcher() -> Agent:
    return Agent(
        config=agents_config["researcher"], verbose=True ,allow_delegation=False, tools=[nl2sql], llm=llm
    )
```

```python
research_agent = researcher()
```

```python
task_define_problem = Task(
    description="""Analyse the data by fetching and executing the correct SQL from the database
            
            Here is the user's problem:

            {topic}
            """.format(topic="I want to get the top 10 articles by RRP"),
    agent=research_agent,
    expected_output="a list of 10 articles formatted in the correct order",
)
```

```python
crew = Crew(
    agents=[research_agent],  # , Summarization_Agent],
    tasks=[task_define_problem],  # , task_summarize],
    verbose=True,
)
```

```python
result = crew.kickoff()
```
